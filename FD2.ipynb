```python
import requests
import time
import pandas as pd
from datetime import datetime, timedelta

# Parameters
symbol = "P-BTC-100000-170525"
resolution = "1h"

end_datetime = datetime(2025, 5, 16, 23, 59)
end_time = int(end_datetime.timestamp())
start_time = end_time - 7 * 24 * 60 * 60  # 7 days back

params = {
    "symbol": symbol,
    "resolution": resolution,
    "start": start_time,
    "end": end_time
}

headers = {
    "Accept": "application/json"
}

# API Call
url = "https://api.india.delta.exchange/v2/history/candles"
response = requests.get(url, params=params, headers=headers)

data=response.json()
print(data)
```

```python
# CONVERTING AND CHECKING THE DATAFRAME
df = pd.DataFrame(data["result"])
print(df.columns)
print(df.head)
```
